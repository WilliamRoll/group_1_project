```python
# dependencies
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

master_csv = "archive/Master.csv"
batting_csv = "archive/Batting.csv"
```

```python
#read in csv files
master_df = pd.read_csv(master_csv)
#master_df.head()
batting_df = pd.read_csv(batting_csv)
batting_df.head()

#filter batting df 
filtered_batting_df = batting_df.loc[batting_df['yearID'] > 1999]
filtered_batting_df
```

```python

merged_df = pd.merge(master_df, filtered_batting_df, on='playerID', how = "inner" )

print(merged_df)
#df.to_csv('df.csv')
```

```python
# drop columns
merged_df = merged_df.drop(['birthDay',
                            'birthMonth',
                            'deathDay',
                            'deathMonth',
                            'deathCountry',
                            'deathState',
                            'deathCity',
                            'weight',
                            'height',
                            'nameFirst',
                            'nameLast',
                            'bats',
                            'throws',
                           ], axis=1)

merged_df
merged_df.dtypes
```

```python
# rename headers to make them neat
merged_df = merged_df.rename(columns={'playerID': 'Player ID',
                                      'birthYear': 'Birth Year',
                                      'birthCountry': 'Birth Country',
                                      'birthState': 'Birth State',
                                      'birthCity': 'Birth City',
                                      'deathYear': 'Death Year',
                                      'nameGiven': 'Full Name',
                                      'weight': 'Weight',
                                      'height': 'Height',
                                      'bats': 'Bats',
                                      'throws': 'Throws',
                                      'debut': 'Debut Date',
                                      'finalGame': 'Date of Last Game',
                                      'retroID': 'Retro ID',
                                      'bbrefID': 'BBRef ID'
                                     })
merged_df['teamID'].nunique()
```

```python
teams_grouped_df = merged_df.groupby('teamID').agg({'AB': ['mean'],'R': ['mean'],'H': ['mean'],'2B': ['mean'],
                                                    '3B': ['mean'],'HR': ['mean'],'RBI': ['mean'],'SB': ['mean'],
                                                    'CS': ['mean'],'BB': ['mean'],'SO': ['mean'],'IBB': ['mean'],
                                                    'HBP': ['mean'],'SH': ['mean'],'SF': ['mean'],'GIDP': ['mean'],
                                                   })
teams_grouped_df.columns = ['At Bat Average', 'Runs Average', 'Hits Average', 'Doubles Average', 'Triples Average',
                           'Home Run Average', 'RBI Average', 'Stolen Bases Average', 'Caught Stealing Average',
                            'Base on Balls Average', 'Strikouts Average', 'Intentional Walks Average',
                            'Hit by Pitch Average', 'Sacrifices Average', 'Sacrifice Flies Average', 
                            'Grounded into Double Plays Average'
                           ]
teams_grouped_df.head()
```

```python
#grouped by teams and sorted average runs of a player on team between 2000-2015
teams_grouped_runs = merged_df.groupby('teamID').agg({'R': ['mean','max','min','std','var']})
teams_grouped_runs.columns = ['Runs Average', 'Runs Max', 'Runs Min', 'Runs STD', 'Runs Var']
print(teams_grouped_runs.reset_index().sort_values(['Runs Average'],ascending=False).set_index(['teamID']).head())
teams_grouped_runs.to_csv('output/Average_Runs_by_Team')
```

```python
#grouped by teams and sorted average stikeout of a player on team between 2000-2015
teams_grouped_SO = merged_df.groupby('teamID').agg({'SO': ['mean','max','min','std','var']})
teams_grouped_SO.columns = ['Strike Out Average', 'Strike Out Max', 'Strike Out Min', 'Strike Out STD', 'Strike Out Var']
print(teams_grouped_SO.reset_index().sort_values(['Strike Out Average'],ascending=False).set_index(['teamID']).head())
teams_grouped_SO.to_csv('output/Average_SO_by_Team')
#teams_grouped_SO['teamID'].nunique()
```

```python
#grouped by teams and sorted average runs of a player on team between 2000-2015
teams_grouped_HR = merged_df.groupby('teamID').agg({'HR': ['mean','max','min','std','var']})
teams_grouped_HR.columns = ['Home Runs Average', 'Home Runs Max', 'Home Runs Min', 'Home Runs STD', 'Home Runs Var']
print(teams_grouped_HR.reset_index().sort_values(['Home Runs Average'],ascending=False).set_index(['teamID']).head())
teams_grouped_HR.to_csv('output/Average_Home_Runs_by_Team')
```

```python
# create column for batting average
#drop columns with 0 at bats
batting_average = merged_df.loc[(merged_df['AB'] !=0)]
#batting_average = merged_df.loc[(merged_df['AB'] != 0).all(axis=1)]
batting_average['Batting Average'] = batting_average["H"]/batting_average["AB"]
#print(batting_average)
grouped_by_teams= batting_average.groupby('teamID').agg({'Batting Average': ['mean','max','min','std','var']})
grouped_by_teams.columns = ['Team Batting Average', 'Batting Average Max', 'Batting Average Min', 'Batting Average STD', 'Batting Average Var']
print(grouped_by_teams.reset_index().sort_values(['Team Batting Average'],ascending=False).set_index(['teamID']).head())
grouped_by_teams.to_csv('output/Batting_Average_by_Team')
```
