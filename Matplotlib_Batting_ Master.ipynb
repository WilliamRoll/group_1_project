```python
import pandas as pd
import matplotlib.pyplot as plt
master_csv = "archive/Master.csv"
batting_csv = "archive/Batting.csv"
#x
```

```python
#read in csv files
master_df = pd.read_csv(master_csv)
#master_df.head()
batting_df = pd.read_csv(batting_csv)
batting_df.head()

#filter batting df 
filtered_batting_df = batting_df.loc[batting_df['yearID'] > 1999]
filtered_batting_df
```

```python
merged_df = pd.merge(master_df, filtered_batting_df, on='playerID', how = "inner" )

print(merged_df)
```

```python
merged_df = merged_df.drop(['birthDay',
                            'birthMonth',
                            'deathDay',
                            'deathMonth',
                            'deathCountry',
                            'deathState',
                            'deathCity',
                            'weight',
                            'height',
                            'nameFirst',
                            'nameLast',
                            'bats',
                            'throws',
                            
                           ], axis=1)

merged_df
```

```python
merged_df = merged_df.rename(columns={'playerID': 'Player ID',
                                      'birthYear': 'Birth Year',
                                      'birthCountry': 'Birth Country',
                                      'birthState': 'Birth State',
                                      'birthCity': 'Birth City',
                                      'deathYear': 'Death Year',
                                      'nameGiven': 'Full Name',
                                      'weight': 'Weight',
                                      'height': 'Height',
                                      'bats': 'Bats',
                                      'throws': 'Throws',
                                      'debut': 'Debut Date',
                                      'finalGame': 'Date of Last Game',
                                      'retroID': 'Retro ID',
                                      'bbrefID': 'BBRef ID'
                                     })
```

```python
teams_grouped_df = merged_df.groupby('teamID').agg({'AB': ['mean'],'R': ['mean'],'H': ['mean'],'2B': ['mean'],
                                                    '3B': ['mean'],'HR': ['mean'],'RBI': ['mean'],'SB': ['mean'],
                                                    'CS': ['mean'],'BB': ['mean'],'SO': ['mean'],'IBB': ['mean'],
                                                    'HBP': ['mean'],'SH': ['mean'],'SF': ['mean'],'GIDP': ['mean'],
                                                   })
teams_grouped_df.head()
#teams_grouped_df.sort_values(by=['HR'])
```

```python
teams_grouped_df = merged_df.groupby('teamID').agg({'AB': ['mean'],'R': ['mean'],'H': ['mean'],'2B': ['mean'],
                                                    '3B': ['mean'],'HR': ['mean'],'RBI': ['mean'],'SB': ['mean'],
                                                    'CS': ['mean'],'BB': ['mean'],'SO': ['mean'],'IBB': ['mean'],
                                                    'HBP': ['mean'],'SH': ['mean'],'SF': ['mean'],'GIDP': ['mean'],
                                                   })
teams_grouped_df.head()
#teams_grouped_df.sort_values(by=['HR'])
```

```python
teams_grouped_runs = merged_df.groupby('teamID').agg({'R': ['mean','max','min']})
teams_grouped_runs.head()
```

```python
#to get mean
dataset_mean = merged_df.groupby('teamID').mean()['H']
#to get median
dataset_median = merged_df.groupby('teamID').median()['H']
#to get variance
dataset_variance = merged_df.groupby('teamID').var()['H']
#to get standard deviation
dataset_stdev = merged_df.groupby('teamID').std()['H']
#to get Standard Error of Mean
dataset_sem = merged_df.groupby('teamID').sem()['H']
#create dataframe for summary
new_df = pd.DataFrame(dataset_mean)
new_df.head()
#rename Tumor Volume column to mean 
new_tables = new_df.rename(columns={"(teamID)": "Mean" })
#median,variance, sem, std for each drug for table
summary_table = new_tables.assign(Median=dataset_median, Variance = dataset_variance,Standard_Error_Mean = dataset_sem , Standard_Deviation = dataset_stdev)
#display  summary stat table
summary_table
```

```python
#grouped by teams and sorted average runs of a player on team between 2000-2015
teams_grouped_runs = merged_df.groupby('teamID').agg({'R': ['mean','max','min','std','var']})
teams_grouped_runs.columns = ['Runs Average', 'Runs Max', 'Runs Min', 'Runs STD', 'Runs Var']
teams_grouped_runs = teams_grouped_runs.reset_index()
teams_grouped_runs.sort_values(by=['Runs Average'],ascending=False)
print(teams_grouped_runs)
teams_grouped_runs.to_csv('output/Average_Runs_by_Team')
```

```python
#create barplot to analyze different teams and their average runs
barplot_dataset = teams_grouped_runs.groupby('teamID').count()['Runs Average']

#create df
barplot_dataset_df = pd.DataFrame(barplot_dataset)
#barplot_dataset_df
plt.xdata = barplot_dataset_df.plot.bar(legend = False, rot = 90, align="center")
plt.title('teamID vs Runs Average')
plt.xlabel('teamID')
plt.ylabel('Runs Average')

#to show results 
plt.show()
#to show results with proper adjustments
plt.tight_layout()
```
